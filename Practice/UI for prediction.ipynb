# UI for final price

```python
import tkinter as tk
from tkinter import StringVar
import joblib
import pandas as pd

# Load the trained model for final price prediction
model = joblib.load('final_price_model.pkl')

# Function to predict the final price based on the current inputs
def predict_final_price(inputs):
    features = ["First Name", "Last Name", "Buyer ID", "Brand", "Model", "Manufacture Country", "Sale Country"]
    
    # Ensure all expected input fields are provided
    missing_fields = set(features) - set(inputs.keys())
    if missing_fields:
        return None
    
    # Create a DataFrame for prediction
    input_df = pd.DataFrame([inputs])
    
    # Make prediction
    prediction = model.predict(input_df)
    return prediction[0]

# Update the final price prediction in real-time
def update_final_price(*args):
    current_inputs = {
        "First Name": first_name_var.get(),
        "Last Name": last_name_var.get(),
        "Buyer ID": buyer_id_var.get(),
        "Brand": brand_var.get(),
        "Model": model_var.get(),
        "Manufacture Country": manufacture_country_var.get(),
        "Sale Country": sale_country_var.get()
    }
    
    # Predict and display the final price
    predicted_final_price = predict_final_price(current_inputs)
    if predicted_final_price is not None:
        final_price_var.set(predicted_final_price)
    else:
        final_price_var.set("")

# Create the main window
root = tk.Tk()
root.title("Final Price Prediction")

# Define StringVars
first_name_var = StringVar()
last_name_var = StringVar()
buyer_id_var = StringVar()
brand_var = StringVar()
model_var = StringVar()
manufacture_country_var = StringVar()
sale_country_var = StringVar()
final_price_var = StringVar()

# Bind the update function to the input variables
first_name_var.trace_add("write", update_final_price)
last_name_var.trace_add("write", update_final_price)
buyer_id_var.trace_add("write", update_final_price)
brand_var.trace_add("write", update_final_price)
model_var.trace_add("write", update_final_price)
manufacture_country_var.trace_add("write", update_final_price)
sale_country_var.trace_add("write", update_final_price)

# Create and place labels and entry widgets
tk.Label(root, text="First Name").grid(row=0, column=0)
tk.Entry(root, textvariable=first_name_var).grid(row=0, column=1)
tk.Label(root, text="Last Name").grid(row=1, column=0)
tk.Entry(root, textvariable=last_name_var).grid(row=1, column=1)
tk.Label(root, text="Buyer ID").grid(row=2, column=0)
tk.Entry(root, textvariable=buyer_id_var).grid(row=2, column=1)
tk.Label(root, text="Brand").grid(row=3, column=0)
tk.Entry(root, textvariable=brand_var).grid(row=3, column=1)
tk.Label(root, text="Model").grid(row=4, column=0)
tk.Entry(root, textvariable=model_var).grid(row=4, column=1)
tk.Label(root, text="Manufacture Country").grid(row=5, column=0)
tk.Entry(root, textvariable=manufacture_country_var).grid(row=5, column=1)
tk.Label(root, text="Sale Country").grid(row=6, column=0)
tk.Entry(root, textvariable=sale_country_var).grid(row=6, column=1)
tk.Label(root, text="Final Price").grid(row=7, column=0)
tk.Entry(root, textvariable=final_price_var, state='readonly').grid(row=7, column=1)

# Run the main event loop
root.mainloop()
```

# UI for combined fields

```python
import tkinter as tk
from tkinter import StringVar
import joblib
import pandas as pd

# Load the trained models and label encoders
last_name_model = joblib.load('using_first_name_for_last_name_model.pkl')
final_price_model = joblib.load('final_price_model.pkl')
label_encoders = joblib.load('label_encoders.pkl')

# Function to handle unseen labels gracefully
def transform_with_encoder(encoder, values):
    unique_labels = encoder.classes_
    transformed_values = []
    for value in values:
        if value in unique_labels:
            transformed_values.append(encoder.transform([value])[0])
        else:
            transformed_values.append(-1)  # Using -1 for unseen labels
    return transformed_values

# Function to predict the last name based on the first name
def predict_last_name(inputs):
    new_df = pd.DataFrame([inputs])
    
    # Encode input features
    for feature in inputs.keys():
        if feature in label_encoders and inputs[feature]:  # Only encode non-empty values
            new_df[feature] = transform_with_encoder(label_encoders[feature], [inputs[feature]])
    
    # Make prediction
    prediction = last_name_model.predict(new_df)
    
    # Decode the prediction
    prediction = label_encoders['Last Name'].inverse_transform(prediction)
    
    return prediction[0]

# Function to predict the final price based on the current inputs
def predict_final_price(inputs):
    features = ["First Name", "Last Name", "Buyer ID", "Brand", "Model", "Manufacture Country", "Sale Country"]
    
    # Ensure all expected input fields are provided
    missing_fields = set(features) - set(inputs.keys())
    if missing_fields:
        return None
    
    # Create a DataFrame for prediction
    input_df = pd.DataFrame([inputs])
    
    # Make prediction
    prediction = final_price_model.predict(input_df)
    return prediction[0]

# Update the last name prediction in real-time
def update_last_name(*args):
    current_inputs = {
        "First Name": first_name_var.get()
    }
    
    # Only consider non-empty inputs for prediction
    non_empty_inputs = {k: v for k, v in current_inputs.items() if v}
    print("Current Inputs for Last Name:", non_empty_inputs)  # Debugging statement

    if "First Name" in non_empty_inputs:
        predicted_last_name = predict_last_name(non_empty_inputs)
        print("Predicted Last Name:", predicted_last_name)  # Debugging statement
        if not last_name_var.get() and predicted_last_name:  # Only update if the field is empty
            last_name_var.set(predicted_last_name)
    update_final_price()

# Update the final price prediction in real-time
def update_final_price(*args):
    current_inputs = {
        "First Name": first_name_var.get(),
        "Last Name": last_name_var.get(),
        "Buyer ID": buyer_id_var.get(),
        "Brand": brand_var.get(),
        "Model": model_var.get(),
        "Manufacture Country": manufacture_country_var.get(),
        "Sale Country": sale_country_var.get()
    }
    
    # Predict and display the final price
    predicted_final_price = predict_final_price(current_inputs)
    if predicted_final_price is not None:
        final_price_var.set(predicted_final_price)
    else:
        final_price_var.set("")

# Clear the last name when first name changes
def first_name_changed(*args):
    last_name_var.set("")
    update_last_name()

# Create the main window
root = tk.Tk()
root.title("Real-Time Prediction Form")

# Define StringVars
first_name_var = StringVar()
last_name_var = StringVar()
buyer_id_var = StringVar()
brand_var = StringVar()
model_var = StringVar()
manufacture_country_var = StringVar()
sale_country_var = StringVar()
final_price_var = StringVar()

# Bind the update functions to the input variables
first_name_var.trace_add("write", first_name_changed)
last_name_var.trace_add("write", update_final_price)
buyer_id_var.trace_add("write", update_final_price)
brand_var.trace_add("write", update_final_price)
model_var.trace_add("write", update_final_price)
manufacture_country_var.trace_add("write", update_final_price)
sale_country_var.trace_add("write", update_final_price)

# Create and place labels and entry widgets
tk.Label(root, text="First Name").grid(row=0, column=0)
tk.Entry(root, textvariable=first_name_var).grid(row=0, column=1)
tk.Label(root, text="Last Name").grid(row=1, column=0)
tk.Entry(root, textvariable=last_name_var).grid(row=1, column=1)
tk.Label(root, text="Buyer ID").grid(row=2, column=0)
tk.Entry(root, textvariable=buyer_id_var).grid(row=2, column=1)
tk.Label(root, text="Brand").grid(row=3, column=0)
tk.Entry(root, textvariable=brand_var).grid(row=3, column=1)
tk.Label(root, text="Model").grid(row=4, column=0)
tk.Entry(root, textvariable=model_var).grid(row=4, column=1)
tk.Label(root, text="Manufacture Country").grid(row=5, column=0)
tk.Entry(root, textvariable=manufacture_country_var).grid(row=5, column=1)
tk.Label(root, text="Sale Country").grid(row=6, column=0)
tk.Entry(root, textvariable=sale_country_var).grid(row=6, column=1)
tk.Label(root, text="Final Price").grid(row=7, column=0)
tk.Entry(root, textvariable=final_price_var, state='readonly').grid(row=7, column=1)

# Run the main event loop
root.mainloop()
```
